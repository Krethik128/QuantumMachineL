```python
from qiskit import QuantumCircuit
qc=QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)
qc.measure([0,1],[0,1])
qc.draw(output='mpl')
```

```python

import pennylane as qml
```

```python


# Define the number of qubits in the circuit
num_qubits = 2

# Create a PennyLane quantum device
dev = qml.device("default.qubit", wires=num_qubits)

# Define the quantum circuit
@qml.qnode(dev)
def my_quantum_circuit(params):
    # Apply a Hadamard gate to the first qubit
    qml.Hadamard(wires=0)
    
    # Apply a Pauli-X (NOT) gate to the second qubit
    qml.PauliX(wires=1)
    
    # Apply a CNOT gate with the first qubit as the control and the second qubit as the target
    qml.CNOT(wires=[0, 1])
    
    # Return the expectation value of the Pauli-Z operator on the second qubit
    return qml.expval(qml.PauliZ(1))

# Define some random parameters
params = [0.1, 0.2]

# Execute the quantum circuit
result = my_quantum_circuit(params)
print("Expectation value of Pauli-Z on qubit 1:", result)
```
