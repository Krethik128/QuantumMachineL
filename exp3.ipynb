```python
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initialize a quantum device with 1 qubit
dev = qml.device('default.qubit', wires=1)

# Quantum circuit with various gates
@qml.qnode(dev)
def circuit():
    qml.PauliX(wires=0)
    qml.PauliY(wires=0)
    qml.PauliZ(wires=0)
    qml.S(wires=0)
    qml.T(wires=0)
    qml.Hadamard(wires=0)
    qml.RX(np.pi/4, wires=0)
    qml.RY(np.pi/4, wires=0)
    qml.RZ(np.pi/4, wires=0)
    return qml.state()

# Using the qml.draw function to visualize the circuit
fig, ax = qml.draw_mpl(circuit)()
plt.show()
result=circuit()
print("Probabilities:",result)


```

```python
# Initialize a quantum device with 2 qubits
dev = qml.device('default.qubit', wires=2)

# Quantum circuit with various gates
@qml.qnode(dev)
def circuit():
    qml.Hadamard(wires=0)
    qml.CNOT(wires=[0,1])  # Apply CNOT gate with qubit 0 as control and qubit 1 as target
    return qml.state()

# Draw the quantum circuit
fig, ax = qml.draw_mpl(circuit)()
plt.show()

# Execute the circuit to obtain the final quantum state
result = circuit()
print("Final quantum state:\n", result)
```
