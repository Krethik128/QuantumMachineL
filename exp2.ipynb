```python
import qiskit 
import numpy 
import math
import matplotlib
```

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram ,circuit_drawer
```

```python
# Create a quantum circuit with 3 qubits and 3 classical bits to demonstrate multi-qubit gates as well
qc = QuantumCircuit(3, 3)

# Apply Pauli gates
qc.x(0)  # Pauli-X on qubit 0
qc.y(1)  # Pauli-Y on qubit 1
qc.z(2)  # Pauli-Z on qubit 2

# Hadamard gate
qc.h(0)  # Apply Hadamard to qubit 0

# Phase gates
qc.s(1)   # S gate on qubit 1
qc.sdg(1) # S dagger on qubit 1
qc.t(2)   # T gate on qubit 2
qc.tdg(2) # T dagger on qubit 2

# CNOT gate
qc.cx(0, 1) # CNOT with qubit 0 as control and 1 as target

# SWAP gate
qc.swap(1, 2) # Swap states of qubit 1 and 2

# Toffoli gate (CCX)
qc.ccx(0, 1, 2) # Toffoli gate with qubits 0 and 1 as controls, and 2 as target

# Measurement
qc.measure([0, 1, 2], [0, 1, 2])
qc.draw(output='mpl')
```
