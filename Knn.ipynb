```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
np.random.seed(0)
```

```python
group1 = np.random.randn(50, 2) + [1, 1]  # Centered around (1, 1)
group2 = np.random.randn(50, 2) + [5, 5]  # Centered around (5, 5)
```

```python
X = np.vstack([group1, group2])
# Generate labels (0 for group 1, 1 for group 2)
y = np.array([0]*50 + [1]*50)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)
```

```python
new_point = np.random.randn(1, 2) + [3, 3]  # Randomly adding near the center
```

```python
prediction = knn.predict(new_point)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(group1[:, 0], group1[:, 1], color='blue', label='Group 1')
plt.scatter(group2[:, 0], group2[:, 1], color='green', label='Group 2')
plt.scatter(new_point[:, 0], new_point[:, 1], color='red', label='New Point', edgecolors='k', s=100)
plt.legend()
plt.title('kNN Classification')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
print("The new point belongs to Group", prediction[0]+1)
```
