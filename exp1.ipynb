```python
import qiskit
import numpy
import matplotlib
import math
```

Bell circuit

```python
from qiskit import QuantumCircuit
qc=QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)
qc.measure([0,1],[0,1])
qc.draw(output='mpl')
```

GHZ circuit

```python
n=3
qc=QuantumCircuit(n,n)
qc.h(0)
for i in range(n-1):
    qc.cx(i,i+1)
qc.measure(range(n),range(n))
qc.draw(output='mpl')
```
